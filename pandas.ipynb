```python
def get_colum_names(schemas, ds_name, sorting_key = 'column_position'):
    column_details = schemas[ds_name]
    columns = sorted(column_details, key =  lambda col : col[sorting_key])
    return [col['column_name'] for col in columns]
```

```python
import json
schemas = json.load(open('schemas.json'))
```

```python
order_items = (schemas, 'order_items')
```

```python
column_name = get_colum_names(schemas ,'order_items')
```

```python
orderitems = pd.read_csv('order_items.txt', names = column_name)
orderitems.head()
```

```python
products_name = (schemas ,'products')
```

```python
column_products = get_colum_names(schemas , 'products')
```

```python
products = pd.read_csv('products.txt',names = column_products)
products.head()
```

```python
orderitems = orderitems.set_index('order_item_product_id')

```

```python
products = products.set_index('product_id')
```

```python
products
```

```python
orderitems
```

```python
join_prod_order = orderitems.join(products, how='inner')
join_prod_order.head()
```

```python
join_prod_order['TotalValue'] = join_prod_order['order_item_quantity'] * join_prod_order['order_item_product_price']
agg = join_prod_order.groupby('order_item_id')['TotalValue'].sum().reset_index()
print(agg.head(10))
```
